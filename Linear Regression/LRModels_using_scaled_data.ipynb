## Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('car_purchasing.csv',encoding='latin-1')
```

```python
df
```

```python
df.describe()
```

```python
df.info()
```

```python
#need to convert the age data type to int
df["age"] = df['age'].astype('int')
```

```python
df
```

```python
df.duplicated().sum()
```

```python
df.corr()
```

```python
#dropping off irrelevant columns
df.drop("gender",axis=1,inplace=True)
```

```python
df.drop(["customer name","customer e-mail"],axis=1,inplace=True)
```

```python
df
```

```python
#plotting subplots to visualize the distribution of data
figure, axis = plt.subplots(2,2,figsize=(10,10))
plt.figure(figsize=(60,40))

axis[0,0].scatter(df['age'],df['car purchase amount'])
axis[0,0].set_title("Age v/s Amount")
axis[0,0].set_xlabel("Age")
axis[0,0].set_ylabel("Amount")

axis[0,1].scatter(df["annual Salary"],df['car purchase amount'])
axis[0,1].set_title("Salary v/s Amount")
axis[0,1].set_xlabel("Salary")
axis[0,1].set_ylabel("Amount")

axis[1,0].scatter(df["credit card debt"],df['car purchase amount'])
axis[1,0].set_title("Debt v/s Amount")
axis[1,0].set_xlabel("Debt")
axis[1,0].set_ylabel("Amount")

axis[1,1].scatter(df["net worth"],df['car purchase amount'])
axis[1,1].set_title("Net Worth v/s Amount")
axis[1,1].set_xlabel("Net Worth")
axis[1,1].set_ylabel("Amount")
```

##Training Model without scaling

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
X = df[["age","annual Salary","credit card debt","net worth"]]
```

```python
X
```

```python
y = df['car purchase amount']
```

```python
y
```

```python
#performing training test split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
m1 = LinearRegression()
```

```python
#training the model
m1.fit(X_train,y_train)
```

```python
m1.coef_
```

```python
m1.intercept_
```

```python
ym1_pred = m1.predict(X_test)
```

```python
r2_score(ym1_pred,y_test)
```

## Applying Standardization

```python
df1 = df.copy()
```

```python
df1
```

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler
```

```python
stdscaler = StandardScaler()
```

```python
#calculating the mean and standard deviation
stdscaler.fit(X_train)
```

```python
stdscaler.mean_
```

```python
# Now transforming the data around the origin.It transforms the dataframe into the numpy array
x_scaled_train1 = stdscaler.transform(X_train)
x_scaled_test1 = stdscaler.transform(X_test)
```

```python
#converting the numpy array into pandas dataframe
x_scaled_train1 = pd.DataFrame(x_scaled_train1,columns=X_train.columns)
x_scaled_test1 = pd.DataFrame(x_scaled_test1,columns=X_test.columns)
```

```python
x_scaled_train1
```

```python
type(x_scaled_train1)
```

```python
np.round(x_scaled_train1.describe(),1)
```

#### Plotting graphs to see if there is change in some kind of data distribution or not

```python
figure, axis = plt.subplots(2,2,figsize=(10,10))
plt.figure(figsize=(60,40))

axis[0,0].scatter(df['age'],df['car purchase amount'])
axis[0,0].set_title("Age v/s Amount (Before Scaling)")
axis[0,0].set_xlabel("Age")
axis[0,0].set_ylabel("Amount")

axis[0,1].scatter(x_scaled_train1['age'],y_train)
axis[0,1].set_title("Age v/s Amount (After Scaling)")
axis[0,1].set_xlabel("Age")
axis[0,1].set_ylabel("Amount")

axis[1,0].scatter(df["annual Salary"],df['car purchase amount'])
axis[1,0].set_title("Salary v/s Amount (Before Scaling)")
axis[1,0].set_xlabel("Salary")
axis[1,0].set_ylabel("Amount")

axis[1,1].scatter(x_scaled_train1["annual Salary"],y_train)
axis[1,1].set_title("Salary v/s Amount (After Scaling)")
axis[1,1].set_xlabel("Salary")
axis[1,1].set_ylabel("Amount")
```

```python
figure2,(ax1,ax2) = plt.subplots(ncols=2,figsize=(10,5))
#before scaling
ax1.set_title("Age Distribution Before Scaling")
sns.kdeplot(X_train['age'],ax=ax1)

#after scaling
ax2.set_title("Age Distribution After Scaling")
sns.kdeplot(x_scaled_train1['age'],ax=ax2)
```

```python
figure3,(ax1,ax2) = plt.subplots(ncols=2,figsize=(10,5))
#before scaling
ax1.set_title("Net Worth Distribution Before Scaling")
sns.kdeplot(X_train['net worth'],ax=ax1)

#after scaling
ax2.set_title("Net Worth Distribution After Scaling")
sns.kdeplot(x_scaled_train1['net worth'],ax=ax2)
```

As you can notice, after performing standardization the distribution of data is the same. It doesn't change

```python
m2 = LinearRegression()
```

```python
#training model on the standardized data
m2.fit(x_scaled_train1,y_train)
```

```python
m2.coef_
```

```python
m2.intercept_
```

```python
ym2_pred = m2.predict(x_scaled_test1)
```

```python
r2_score(ym2_pred,y_test)
```
